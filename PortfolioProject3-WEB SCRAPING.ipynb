```python
#Import libraries
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
import pandas as pd
import numpy as np
import requests
import time
import datetime
import smtplib #helps send email to self
```

```python
#Appending the url of all the individual oages into one combines list
URL = "https://www.flipkart.com/search?q=apple%20iphone%2013&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page=1"
urls = []
#in url done i"%(var) 
for var in range(1,9):
    url = ["https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page=%i"%(var)]
    urls.append(url[0])
urls#stores info of all the urls
```

```python
def connect(url):    
    uClient = urlopen(url)
    page_html = uClient.read()
    uClient.close()
    #fetch all data in html format and store above data in a variable
    page_soup = soup(page_html, "html.parser")
    container = page_soup.find_all("div",{"class": "_2kHMtA"})
    return container
```

```python
#Checking if the code works for 1 product in page
con = connect(urls[0])
print(soup.prettify(con[1]))
```

```python
#applied for getting results from whole page for all 8 result pages
for i in range(0,8):
    con = connect(urls[i])
    for container in con:
        product_name = container.div.img["alt"]
        print(product_name)
```

```python
#Cases occurs where actual price and selling price is same, so we us exception handling to cater such cases
def dataframe(containers):
    #Create a dictionary and append it one by one
    data = {"Product_name":[],"Sale price":[],"Actual Price":[],"Rating":[],"Person Rating":[],"Offer":[]}
    for i in range(0,8):
        con = connect(containers[i])
        for container in con:
            #Product Name
            product_name = container.div.img["alt"]
            data["Product_name"].append(product_name)
            
            #Sale price
            s_price = container.findAll("div",{"class":"_30jeq3 _1_WHN1"})
            s_p = s_price[0].text
            data["Sale price"].append(s_p)
            
            #Actual Price
            try:
                actual_price = container.findAll("div",{"class":"_3I9_wc _27UcVY"})
                a_p = actual_price[0].text
                data["Actual Price"].append(a_p)
            except IndexError:
                print("Same as sale price")
                data["Actual Price"].append("Same Price")
            
            #Rating
            try:
                rating = container.findAll("div",{"class":"_3LWZlK"})
                r = rating[0].text
                data["Rating"].append(r)
            except IndexError:
                data["Rating"].append("No Rating")
            
            #No of person given rating
            try:
                p_rating = container.findAll("span",{"class":"_2_R_DZ"})
                p_r = p_rating[0].text
                data["Person Rating"].append(p_r)
            except IndexError:
                data["Person Rating"].append("No Rating")
                        
            
            #Offer
            try:
                offer = container.findAll("div",{"class":"_3Ay6Sb"})
                off = offer[0].text
                data["Offer"].append(off)
            except IndexError:
                data["Offer"].append("No offer this product")
            
        print("done",i) #to show which age is running 
        
    return data

```

```python
dataset = dataframe(urls)
```

```python
df = pd.DataFrame(dataset)
df.dtypes
```

```python
#Rating shoul be numeric ie floating point
df["Rating"] = pd.to_numeric(df["Rating"])
df.dtypes
```

```python
#Separate number of rating and number of reviews
df[["No of Rating","No of Reviews"]] = df["Person Rating"].str.split("&",1,expand=True)
```

```python
#Not have the work "Rating" at end for the number of rating and number of reviews
df["No of Rating"] = df["No of Rating"].str.split(" ").str[0]
```

```python
df["No of Reviews"] = df["No of Reviews"].str.split(" ").str[0]
```

```python
# Code used to drop Person Rating and No. of Reviews when it existed in the table, axis = 1 means removing from columns, 0 for rows
#df = df.drop(["Person Rating","No of Reviews"],axis=1)
df = df.drop(["Person Rating"],axis=1)
df
```
